```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open("embed_run/loss_by_epoch_D_descperimg_5_stopwords_0_beta1_0.300000_lr_0.000400.out", 'r') as file:
    d_loss_values = [float(line) for line in file.readlines()]
    
with open("embed_run/loss_by_epoch_G_descperimg_5_stopwords_0_beta1_0.300000_lr_0.000400.out", 'r') as file:
    g_loss_values = [float(line) for line in file.readlines()]
```

```python
plt.figure()
plt.plot(d_loss_values, color='purple',
         lw=2.5,label="discriminator")
plt.plot(g_loss_values, color='darkgreen',
         lw=2.5, label="generator")
plt.xlim([0, 100])
plt.ylim([0.0, 40])
plt.xlabel('epochs')
plt.ylabel('loss')
plt.title('Word2Vec without stopwords')
plt.grid(True)
plt.legend(loc="upper left")
plt.show()
```
